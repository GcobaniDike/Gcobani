```python
import matplotlib.pyplot as plt

data = {'Coke': 24, 'Fanta': 38, 'Twizza': 52, 'Stoney': 62}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('COOLDRINKS')
```

```python
import matplotlib.pyplot as plt
import numpy as np

choc_name = ["Cadbury", "Tex", "P.S", "Lunch Bar", "Chomp"]
choc_price = [55, 62, 50, 49, 56 ]
choc_quantity = [30, 45, 42, 44, 48]

#create scatter plot
fig, ax = plt.subplots()
ax.scatter(choc_price,choc_quantity,s=choc_quantity,marker='o', c=choc_price)

#label each bubble
for i in range(len(choc_price)):
    plt.annotate("({}) {}".format(i+1,choc_name[i]),xy=(choc_price[i], choc_quantity[i]))

#Label axis
plt.xlabel('CHOC QUANTITY')
plt.ylabel(' CHOC PRICE')
plt.title('CHOCOLATES')
plt.show()
```
